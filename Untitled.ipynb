## Principal Component of Analysis

```python
from sklearn.decomposition import PCA
```

```python
df_scaled_df.isnull().sum()
```

```python
df_scaled_df.dropna(inplace=True)
```

```python
df_scaled_df.describe()
```

```python
pca = PCA()
pca.fit(df_scaled_df)
# data_pca = pca.transform(df_scaled_df)
```

```python
pca.components_
```

Screeplot ------ Explains which component captured the whole variance(the correct info. of the populated data)

```python
explained_variance = pd.DataFrame(pca.explained_variance_)
```

```python
ax = explained_variance.plot.bar()
ax.set_xlabel('PCA components')
ax.set_ylabel('Eigen Values')
```

```python
explained_variance_ratio = pd.DataFrame(pca.explained_variance_ratio_)
```

```python
explained_variance_ratio
```

```python
cumsum = np.cumsum(np.round(explained_variance_ratio,2))
cumsum
```

```python
fig, ax = plt.subplots(figsize=(4,4))
ax.set(title='Explained Variance Ratio')
x = list(range(explained_variance.shape[0]))
ax.plot(x,explained_variance_ratio,marker='o',color='blue',label='ratio')
ax.plot(x,cumsum,marker='*',color='red',label='cumsum ratio')
plt.legend()
plt.show()
```

## KMeans Clustering

```python
from sklearn.cluster import KMeans
```

```python
inertia = []
for n_cluster in range(1,11):
    kmeans = KMeans(n_clusters=n_cluster, random_state=0).fit(df1_scaled_df)
    inertia.append(kmeans.inertia_/n_cluster)
```

```python
inertias = pd.DataFrame({'n_clusters':list(range(1,11)),'Inertia':inertia})
```

```python
fig, ax = plt.subplots(figsize=(10,10))
ax.plot(inertias['n_clusters'],inertias['Inertia'],marker='o')
ax.set(xlabel='no of clusters',ylabel='Inertia\'s',title='UnNormalize Data -- Elbow Method',xticks=range(1,11))
plt.ylim((0,1.1*inertias["Inertia"].max()))
plt.show()
```

```python
from yellowbrick.cluster import KElbowVisualizer
```

```python
model = KMeans()
visualizer = KElbowVisualizer(model, k=(1,10))
visualizer.fit(df1_scaled_df)
visualizer.poof()
```

```python
kmeans1 = KMeans(n_clusters=3)
kmeans1.fit(df_scaled_df)
```

```python
labels1 = kmeans1.labels_
labels1
```

```python
from sklearn.metrics import silhouette_score
silhouette_score(df_scaled_df,labels=labels1)
```
